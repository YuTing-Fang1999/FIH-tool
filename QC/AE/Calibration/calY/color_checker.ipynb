```python
import cv2
import numpy as np
from colour_checker_detection import detect_colour_checkers_segmentation
import matplotlib.pyplot as plt

def read_img(img_path):
    # to BGR to RGB
    img = cv2.imread(img_path)
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)/255
    return img
    
img_my = read_img('my.jpg')
img_ref = read_img('ref.jpg')

```

```python
plt.imshow(img_my)
plt.axis('off')  # Remove axis
plt.show()
```

```python
def pixel_to_luma(pixel):
    luma = 0.299*pixel[0] + 0.587*pixel[1] + 0.114*pixel[2]
    return luma*255
```

```python
colour_checker_swatches_data = detect_colour_checkers_segmentation(img_my, additional_data=True)[0]
swatch_colours, colour_checker_image, swatch_masks = (colour_checker_swatches_data.values)
print(pixel_to_luma(swatch_colours[18]))
print(pixel_to_luma(swatch_colours[19]))
swatch_colours = np.array(swatch_colours).reshape(1, 24, 3)
plt.imshow(swatch_colours)
plt.axis('off')  # Remove axis
plt.show()

# Using the additional data to plot the colour checker and masks.
masks_i = np.zeros(colour_checker_image.shape)
for i, mask in enumerate(swatch_masks):
    masks_i[mask[0]:mask[1], mask[2]:mask[3], ...] = 1
    
plt.imshow(np.clip(colour_checker_image + masks_i * 0.25, 0, 1))
plt.axis('off')  # Remove axis
plt.show()
```
